```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Last year's revenue
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations) #numpy's random normal distribution generator using the arguments above
rev
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
COGS = -(rev * np.random.normal(0.6,0.1))  #assume COGS amount to 60% of revenues

plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
GP = rev + COGS
GP

plt.figure(figsize=(15, 6))
plt.plot(GP)
plt.show()
```
