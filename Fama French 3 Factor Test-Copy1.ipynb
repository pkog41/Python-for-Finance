```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_csv(r"C:\Users\patrickogrady\FFtest.csv" ,parse_dates=True,index_col=0)

```

```python
scatterplotmatrix = sns.pairplot(df)

scatterplotmatrix.fig.set_size_inches(15, 15)

plt.show()
```

```python
from statsmodels.formula.api import ols

x1 = df['Mkt-RF'].tolist()
x2 = df['SMB'].tolist()
x3 = df['HML'].tolist()
y = df['RF'].tolist()

model = ols("y ~ x1+x2+x3", data=df).fit()

print(model.summary())
```

```python
import statsmodels.api as sm

fig = sm.graphics.plot_partregress_grid(model)

fig.set_size_inches(18.5, 10.5)

fig.tight_layout(pad=1.0)

plt.show()

fig = sm.graphics.plot_ccpr_grid(model)

fig.set_size_inches(18.5, 10.5)

fig.tight_layout(pad=1.0)

plt.show()
```

```python
res=model.resid
QQ = sm.qqplot(res)
QQ.set_size_inches(18.5, 10.5)
plt.show()
```
